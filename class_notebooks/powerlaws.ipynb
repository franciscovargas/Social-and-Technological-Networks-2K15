```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
g = nx.erdos_renyi_graph(500,0.2) #random graph
```

```python
nx.draw(g)
```

```python
ds = nx.degree(g).values()
```

```python
plt.hist(ds)
```

```python
web = nx.read_edgelist("web-Google.txt.gz") 
#you can download this file from https://snap.stanford.edu/data/index.html
```

```python

webdeg = nx.degree(web).values()
```

```python
plt.hist(webdeg,bins=50);
```

```python
plt.hist(webdeg,bins=50)
plt.axis([0,1000,0,3000])
```

```python
c = plt.hist(webdeg,bins=40)
```

```python
from collections import Counter
```

```python
c = Counter(webdeg)
```

```python
plt.loglog(c.keys(),c.values(), 'o');
#plt.axis([0,1000,0,3000])
```

```python
pg = nx.barabasi_albert_graph(800,3) #preferential attachment graph
```

```python
pgdegrees = nx.degree(pg).values()
plt.hist(pgdegrees)
```

```python
c = Counter(pgdegrees)
plt.loglog(c.keys(),c.values(), 'o')
```
